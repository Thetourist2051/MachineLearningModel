```python
def percent_missing(df):
    percent_nan = 100*df.isnull().sum()/len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()

    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
percent_nan
```

```python
sns.barplot(x=percent_nan.index, y= percent_nan)
plt.xticks(rotation=90)
plt.ylim(0,1)
plt.show()
```

```python
def miss_per(df):
    null_per = 100*df.isnull().sum()/len(df)
    null_per = null_per[null_per>0].sort_values()
    return null_per
```

```python
per_missing =  miss_per(df)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=per_missing.index, y=per_missing)
plt.ylim(0,1)
plt.xticks(rotation=90);
```
